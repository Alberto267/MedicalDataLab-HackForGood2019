# Resonancias Magnéticas: imagen cardíaca
----

En este Jupyter Notebook se procederá a una primera toma de contacto con los archivos que contienen las imágenes de las resonancias magnéticas. Las imágenes corresponden a un conjunto de datos abierto que se presentó para el **Automatic Cardiac Diagnosis Challenge (ACDC)** que tuvo lugar en el **MICCAI 2017**, uno de los congresos más importantes de imagén médica y Deep Learning. Para el desarrollo del mismo, se dispone de un conjunto de entrenamiento `train`, con 100 pacientes diagnosticados; y un conjunto `test`, con 50 pacientes sin diagnosticar.

Para más información tanto del reto como de los datos: [ACDC Database](https://www.creatis.insa-lyon.fr/Challenge/acdc/databases.html)

```python
import os
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
def show_slices(slices):
    plt.figure(figsize=(40,20))
    fig, axes = plt.subplots(1, len(slices))
    for i, slice in enumerate(slices):
        axes[i].imshow(slice.T, cmap="gray", origin="lower")
```

### Lectura de Imágenes

```python
path_patient_ori = "/home/ubuntu/ACDC_yeah/acdc_data/train/patient001/" # Patient 1 (example)
path_raw = path_patient_ori + 'patient001_4d.nii.gz' 
path_gt  = path_patient_ori + 'patient001_ED_gt.nii.gz'

raw_image = nib.load(path_raw)
gt_image  = nib.load(path_gt)

raw_array = raw_image.get_fdata()[:,:,:,0]
gt_array  = gt_image.get_fdata()
```

### Vista de imágenes

```python
for i in range(10):

    fig, axis = plt.subplots(1, 2, figsize = (5, 5))
    axis[0].imshow(raw_array[:,:,i], cmap = "bone")
    axis[1].imshow(gt_array[:,:,i], cmap = "bone")

plt.show()
```
